```python
# Functions- scope
# Define function, and define varieable outside function but before it's called

def greet_person(l):
    print(f"Hello {l}")
# l is a parameter = which is a vairable available inside this function only and used in the print statment

name = "Kristen" 
# value of variable "name" is passed into function as a parameter
greet_person(name)
# will use name vairable as l parameter

# passing in a new value as a parameter into function. Value only works in this function. Not a variable stored anywhere 
greet_person('Jon')

# greet_person() run by itself raises an exception, because <TypeError: greet_person() missing 1 required positional argument: 'l'>


def greet_person2(): 
    print(f"Hello {k}")
k = "Cappy" 
greet_person2()
# if you dont define any parameters within the function definitation, 
# when called it will look for variable (within scope) to use for print function

# bad form to use by itself, mostly used for nested nested functions, if at all


def greet_persons(*j):
    for i in j:
        print(f"Hello {i}")
# *j is a variable length tuple, allows an arbitrary number or arguements to the function
# if you want to mutate the variable being passed into function, created a new variable insdie function 

greet_persons("Baloo", "Oppie")

# looknig ahead-- functions dont have nay parameters that you don't assign.
# --methods have parameters
```
